```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
s = pd.Series([1.2, 0.1, 1.8, 0.3], index=["am hi", "am lo", "pm hi", "pm lo"])
train_data = pd.read_csv(os.path.join(os.path.split(os.getcwd())[0], "SYD_DAT_5/data/sydtrains.csv"), index_col=1, thousands=',')
```

```python
s.plot(kind='line');
plt.xlabel("Time of Tide");
plt.ylabel("Height of Tide");
```

```python
cbd2014 = train_data[(train_data.LINE == "CBD") & (train_data.YEAR == 2014)]
```

```python
cbd2014[['OUT_0600_0930', 'IN_1500_1830']].plot(kind='bar')
```

```python
cbd2014.IN_24_HOURS.plot(kind='bar',
    color=['red' if name=='Circular Quay' or name=='Museum' else 'blue' for name in cbd2014.index]
                        )
```

```python
np.where(cbd2014.IN_0200_0600 < 1000, "blue", "red")
```

```python
cbd2014.IN_0200_0600.plot(kind='bar',
                         color=np.where(cbd2014.IN_0200_0600 < 500, "blue", "red"))
```
